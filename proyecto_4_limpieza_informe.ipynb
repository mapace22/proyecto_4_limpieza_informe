```python
# Usa las tablas department y aisle para revisar los datos del pasillo con ID 100 y el departamento con ID 21.

def verificar_aisle_department(products, departments, aisles, aisle_id=100, department_id=21):
    """
    Revisar datos de aisle_id y department_id en el DataFrame products.

    Args:
        products (pd.DataFrame): DataFrame de info productos.
        departments (pd.DataFrame): DataFrame de info departamentos.
        aisles (pd.DataFrame): DataFrame de info pasillos.
        aisle_id (int): aisle_id por relacionar.
        department_id (int): department_id por relacionar.

    Returns:
        pd.DataFrame: DataFrame / productos que cumplen los criterios.
    """

    # Filtro productos / aisle_id y department_id
    productos_filtrados = products[(products['aisle_id'] == aisle_id) & (products['department_id'] == department_id)]

    # Agregar nombres / aisle y department
    productos_filtrados = pd.merge(productos_filtrados, aisles[['aisle_id', 'aisle']], on='aisle_id', how='left')
    productos_filtrados = pd.merge(productos_filtrados, departments[['department_id', 'department']], on='department_id', how='left')

    return productos_filtrados

# Comprobar datos / aisle_id 100 y department_id 21
productos_verificados = verificar_aisle_department(products, departments, aisles)

# Mostrar
if not productos_verificados.empty:
    print(f"Productos con aisle_id 100 y department_id 21:")
    print(productos_verificados)
else:
    print(f"No se encontraron productos con aisle_id 100 y department_id 21.")
```

### [C1] ¿Cuántos artículos compran normalmente las personas en un pedido? ¿Cómo es la distribución?

```python

# Contar la cantidad de artículos por pedido
items_per_order = order_products.groupby('order_id').size().reset_index(name='num_items')

# Distribución estadística
mean_items= items_per_order ['num_items'].mean()
median_items= items_per_order ['num_items'].median()
mode_items= items_per_order ['num_items'].mode().values
min_items= items_per_order ['num_items'].min()
max_items= items_per_order ['num_items'].max()

#Print datos estadísticos
print("Estadísticas sobre el # de artículos por pedido")
print(f"- Media: {mean_items:.2f}")
print(f"- Mediana: {median_items}")
print(f"- Moda: {mode_items}")
print(f"- Mínimo: {min_items}")
print(f"- Máximo: {max_items}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np

# Gráfico
plt.figure(figsize=(12, 6))
sns.histplot(items_per_order['num_items'], bins=30, color='skyblue', edgecolor='black', alpha=0.7, kde=True)


plt.title('Distribución del Número de Artículos por Pedido', fontsize=16)
plt.xlabel('Número de Artículos por Pedido', fontsize=12)
plt.ylabel('Densidad', fontsize=12)
plt.grid(True)

# Añadir líneas verticales para la media, mediana y moda
plt.axvline(x=mean_items, color='red', linestyle='--', label=f'Media: {mean_items:.2f}')
plt.axvline(x=median_items, color='green', linestyle='--', label=f'Mediana: {median_items}')
plt.axvline(x=mode_items, color='orange', linestyle='--', label=f'Moda: {mode_items}')

# Mostrar leyenda
plt.legend()

# Mostrar el gráfico
plt.show()
```

### [C3] Para cada producto, ¿cuál es la proporción de las veces que se pide y que se vuelve a pedir?

```python
#Total de pedidos por producto
total_orders = order_products.groupby('product_id').size().reset_index(name='total_orders')

#Total reordenaciones
reordered_orders = order_products[order_products['reordered']==1]
reordered_count = reordered_orders.groupby('product_id').size().reset_index(name='reordered_orders')

#Combinar pedidos y reordenaciones
product_stats = pd.merge(total_orders, reordered_count, on='product_id', how='left')

# Manejar valores NaN de  productos no reordenados
product_stats['reordered_orders'] = product_stats['reordered_orders'].fillna(0)

# Proporción de reordenación
product_stats ['reorder_proportion']= product_stats['reordered_orders']/product_stats['total_orders']

# Unir con tabla de productos para conseguir nombres
product_stats= pd.merge(product_stats, products, on='product_id', how='left')

# Filtrar con total_orders > 1 para evitar proporciones confusas
product_stats = product_stats[product_stats['total_orders'] > 1]

#ordenar por proporción
product_stats= product_stats.sort_values(by=['total_orders', 'reorder_proportion'], ascending=[False, False])

print(product_stats[['product_id','product_name','total_orders','reordered_orders','reorder_proportion']])
```

Es de esperar que los productos que además de poseer mayor popularidad y que se reordenan mayormente, sean los que poseen la mejor proporción. En este screenshoot se muestra más claramente: 


- BANANA: 
  - Se pidió 66.050 veces
  - Se reordenó 55.763 veces
  - Proporción de reordenación (55.763 / 66.050 = 84.4%.)
- BAG OF ORGANIC BANANAS:
  - Se pidió 53.297 veces
  - Se reordenó 44.450 veces
  - Proporción de reordenación (44.450 / 53.297 = 83.4%.)


✅ Para calcular mejor debido a que no se muestra la totalidad de las filas se agregan los siguientes 2 gráficos:

```python
import pandas as pd

# Columna 'total_orders_group'
product_stats['total_orders_group'] = pd.cut(
    product_stats['total_orders'],
    bins=[0, 100, 500, 1000, float('inf')],
    labels=['0-100', '101-500', '501-1000', '1001+']
)

# Columna 'reorder_proportion_group'
product_stats['reorder_proportion_group'] = pd.cut(
    product_stats['reorder_proportion'],
    bins=[0, 0.25, 0.50, 0.75, 1.00],  # Define los rangos
    labels=['0-0.25', '0.26-0.50', '0.51-0.75', '0.76-1.00']
)


summary_total_orders = product_stats.groupby('total_orders_group').size().reset_index(name='count')
print(summary_total_orders)

# Crear el gráfico
plt.figure(figsize=(10, 6))
summary_total_orders.plot(kind='bar', x='total_orders_group', y='count', legend=False, color='skyblue')

# Personalizar el gráfico
plt.title('Distribución de Productos por Rango de Pedidos', fontsize=16)
plt.xlabel('Rango de Pedidos', fontsize=12)
plt.ylabel('Número de Productos', fontsize=12)
plt.grid(True)

# Mostrar el gráfico
plt.show()


print("""

🔍Conclusiónes del gráfico:

   💡Productos de pocos pedidos (0-100):

       -34,047 productos se pidieron entre 0 y 100 veces, son la mayoría. 
       -Tienen bajo volumen de ventas (nuevos, de nicho o poco populares)

    💡Productos de pedidos moderados (101-500):

       -4,879 productos se pidieron entre 101 y 500 veces.
       -Volumen de ventas moderado (promocionarlos más y mejorar su visibilidad).

    💡Productos de muchos pedidos (501-1000 y 1001+):
       - 862 productos se pidieron entre 501-1000 y 738 más de 1000 pedidos.
       - Más populares y prioritarios para gestión de inventario y promoción

""")
```
